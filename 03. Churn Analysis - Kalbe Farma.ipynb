```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

## Library visualisasi
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Kalbe Farma Clean
df=pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQOZTwOzhKxgSO8R2ujHOJJ2deDTYq8uiadDZnJ3tMjybA2StIBsr_33Drtkrwx1w/pub?output=csv')
df.head()
```

```python
# Pastikan kolom transaction_date dalam format datetime
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
```

```python
print(df['transaction_date'].dtype)
```

```python
# Menambahkan fitur churn ke dalam dataset df
df["last_transaction"] = df.groupby("customer_id")["transaction_date"].transform("max")

# Menentukan tanggal terakhir dalam dataset
last_date = df["transaction_date"].max()

# Menandai pelanggan yang tidak memiliki transaksi dalam 90 hari setelah transaksi terakhir mereka sebagai churn (1 = churn, 0 = tidak churn)
df["churn"] = (df["last_transaction"] < (last_date - pd.Timedelta(days=90))).astype(int)

# Hapus kolom sementara last_transaction
df.drop(columns=["last_transaction"], inplace=True)

# Menampilkan hasil yang diperbarui
df.head()
```

```python
# Simpan DataFrame ke file Excel
df.to_excel('Kalbe Churn.xlsx', index=False)
# Menampilkan notifikasi bahwa file telah tersimpan
print("File 'Kalbe Churn.xlsx' telah berhasil disimpan.")
```
